```pyspark
from pyspark.sql import SparkSession
import time
import csv
import io

spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext

print(f"Executors: {sc.getConf().get('spark.executor.instances')}")
print(f"Master: {sc.master}")
```

```pyspark
# RDD
crimes_old = sc.textFile("s3://initial-notebook-data-bucket-dblab-905418150721/project_data/LA_Crime_Data/LA_Crime_Data_2010_2019.csv")
crimes_new = sc.textFile("s3://initial-notebook-data-bucket-dblab-905418150721/project_data/LA_Crime_Data/LA_Crime_Data_2020_2025.csv")
mocodes = sc.textFile("s3://initial-notebook-data-bucket-dblab-905418150721/project_data/MO_codes.txt")
```

```pyspark
# Filter and join data
header_old = crimes_old.first()
header_new = crimes_new.first()

crimes_old_clean = crimes_old.filter(lambda line: line != header_old)
crimes_new_clean = crimes_new.filter(lambda line: line != header_new)
crimes_rdd = crimes_old_clean.union(crimes_new_clean)

crimes_rdd.count()
```

```pyspark
def extract_mocodes(row_string):
    reader = csv.reader(io.StringIO(row_string))
    fields = next(reader)

    mocode_field = fields[10]
    if not mocode_field:
        return []
    codes = mocode_field.split(' ')
    return [(c, 1) for c in codes]
```

```pyspark
times = []
runs = 10

for i in range(runs):
    mocode_pairs = mocodes.filter(lambda x: len(x) > 0) \
            .map(lambda line: line.split(' ', 1)) \
            .filter(lambda x: len(x) == 2) \
            .map(lambda x: (x[0].strip(), x[1].strip()))

    mocode_counts = crimes_rdd.flatMap(extract_mocodes) \
                              .reduceByKey(lambda a, b: a + b)

    mocodes_joined = mocode_counts.join(mocode_pairs)

    result = mocodes_joined.map(lambda x: (x[0], x[1][1], x[1][0])) \
                           .sortBy(lambda x: x[2], ascending=False)

    start = time.time()
    result.count()
    end = time.time()
    times.append(end-start)

for i, t in enumerate(times):
    print(f"Run {i+1}: {t} sec")
rdd_mean = sum(times)/runs
print(f"Mean execution time for RDDs: {rdd_mean}")
```
